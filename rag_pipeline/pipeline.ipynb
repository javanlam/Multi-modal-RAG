## Demonstration of RAG pipeline
Example documents from CLIP-based model testing used as knowledge base

```python
from config.settings import RAGConfig
from main import RAGSystem
```

```python
config = RAGConfig(
    chunk_size=800,
    top_k=5,
    llm_provider="openai-azure"
)

rag = RAGSystem(config)
```

Comparison with generation without the implemented RAG pipeline

```python
import dotenv

dotenv.load_dotenv()
import os

from openai import AzureOpenAI

api_key = os.getenv("OPENAI_API_KEY")

openai_client = AzureOpenAI(
    api_key=api_key,
    api_version="2024-06-01",
    azure_endpoint="https://hkust.azure-api.net",
    azure_deployment="gpt-4o-mini",
)
```

```python
response = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": "What transportation options are available to HKUST?"}
    ],
)

print(response.choices[0].message.content)
```
